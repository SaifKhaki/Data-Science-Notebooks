# Configuration

```python
sns.set_theme()
sns.set_style('darkgrid')
```

Set_style has four arguments:

- darkgrid (default of set_theme)
- white
- whitegrid
- ticks

```python
sns.set_context('talk')
```

set_context has three arguments:
- paper (for just bare look at data)
- talk (more detailed)
- poster (more zoom but less detailed)

# Parameteres to remeber

- data: the df
- kind: empty | axis_level_functions i.e. scatter, line, hist, kde etc
- x: x parameter
- y: y parameter
- col: separating plots on the basis of which parameter
- hue: colors of points separating otbo which parameter
- style: changing style of plot lines otbo which parameter
- size: changing size/width of plots otbo which parameter
- kde: boolean, kernel density estimation i.e. the line estimating histograms
- rugs: boolean, small lines on x axis showing upper class boundary
- split: splitting of violin otbo a parameter
- multiple: ['layer', 'stack', 'fill', 'dodge']
- marker: markers in scatter plot, 'X' | any 
- dodge: boolean, makes pointplot more understandable by separating the lines

### Scatter Plot

```python
sns.relplot(
    data=df,
    x="Age", y="Fare", col="Sex",
    hue="Survived", style="family", size="Pclass"
)
```

This relplot is actually a relationship plot which is actually plotting the relationship between 6 parameters.
but if you can focus we cant really have a better understanding of the data although more relationship of the data under a single plot.

### Linear regression model Line

```python
sns.lmplot(
    data=df,
    x='Age',y='Fare',col='Sex',
    hue='Survived'
)
```

### KDE & Multiple example - histogram

```python
sns.displot(
    data=df, 
    x="Age", 
    hue= 'Survived',
    kde=True, multiple='stack'
)
```

### Difference betweem Stack and Col

```python
sns.displot(
    data=df, 
    x="Age", col= 'Survived', 
    hue='Survived', 
    kde=True
)
```

this is what depicted in the above image but just in one plot.

### Rug example - ECDF Histogram

```python
sns.displot(
    data=df, kind='ecdf', 
    x="Age", 
    rug=True
)
```

### Splilt example - Violin Categorical plot

```python
sns.catplot(data=df, kind="violin", x="Pclass", y="Fare", hue="Sex", split=False)
```

```python
sns.catplot(data=df, kind="violin", x="Pclass", y="Fare", hue="Sex", split=True)
```

both are pretty informative, but in the split plot, we have a hybrid form of internal box plot i.e. in non-split plot, we can clearly identify all the quantiles as well as IQR range along with outlier range. 

### Dodge example - Point Plot and Bar Plot

```python
sns.set_context('paper')
sns.pointplot(
    data=df, 
    x="Pclass", y="Fare", 
    hue="Sex",
    dodge=True
)
```

```python
sns.barplot(
    data=df, 
    x="Pclass", y="Fare", 
    hue="Sex"
)
```

both are pretty same, just shading the bottom area from the central tendency in barplot. However, as we can see the point plot is more useful.
